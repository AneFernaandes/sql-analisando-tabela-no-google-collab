#  Coletando dados do banco de dados
 1. Selecione as colunas: loyalty_number, year, month, flights_booked, total_flights, distance e points_accumulated da tabela ‚Äúflight_activity‚Äù

```python

#sempre utilizar 3 aspas para usar o SELECT FROM WHERE no Collab
consulta_atividade= """

    SELECT
    loyalty_number,
    year,
    month,
    flights_booked,
    total_flights,
    distance,
    points_accumulated


    FROM
    flight_activity


"""
df_atividade = pd.read_sql_query(consulta_atividade, conn)

df_atividade.head()

```

  2.  Selecione as mesmas colunas, por√©m, recupere somente as linhas cuja coluna distance √© maior do que 2000, da tabela‚Äúflight_activity‚Äù

```python

consulta_atividade = """

  SELECT
    fa.loyalty_number,
    fa.year,
    fa.month,
    fa.flights_booked,
    fa.total_flights,
    fa.distance,
    fa.points_accumulated

  FROM
    flight_activity fa
  WHERE
    distance < 2000

 """
df_atividade = pd.read_sql_query(consulta_atividade, conn)
df_atividade.head()
```

3. Selecione as colunas: loyalty_number, month, year, distance, flights_booked e total_flights da tabela ‚Äúflight_activity‚Äù, e
recupere somente as linhas cuja coluna ‚Äúflights_booked‚Äù √© maior do que 10 e a coluna ‚Äúflights_booked‚Äù √© menor do que 12, da tabela ‚Äúflight_activity‚Äù

```python
consulta_atividade = """

  SELECT
    fa.loyalty_number,
    fa.month,
    fa.year,
    fa.distance,
    fa.flights_booked,
    fa.total_flights


  FROM
    flight_activity fa
  WHERE
    fa.flights_booked > 10 AND fa.flights_booked < 12


 """

df_atividade = pd.read_sql_query(consulta_atividade, conn)
df_atividade.head()
```
